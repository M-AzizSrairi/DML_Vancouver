**Correlation Matrix**

*Counts*

```python
import pandas as pd

# Load your data
# Replace 'your_data.csv' with your actual data file path
data_count = pd.read_excel('data\\final_dataset_count.xlsx')

filtered_columns = ['crime_count', 'elevation', 'scaled_betweenness_centrality', 'store_count', 'lockdown', 'graffiti_count', 
                    'homeless_shelter_count', 'traffic_signal_count', 'street_lighting_poles_count', 
                    'police_count', 'transit_count']

filtered_data_count = data_count[filtered_columns]

# Compute the correlation matrix
corr_matrix = filtered_data_count.corr()

# Select the rows and columns
selected_rows = ['crime_count', 'transit_count']  # Adjust these to your actual column names
selected_columns = ['store_count', 'graffiti_count', 
                    'homeless_shelter_count', 'traffic_signal_count', 'street_lighting_poles_count', 
                    'police_count', 'transit_count']

filtered_matrix_count = corr_matrix.loc[selected_rows, selected_columns]

# Display the filtered correlation matrix
print(filtered_matrix_count)

```

*Reach*

```python
# Load your data
# Replace 'your_data.csv' with your actual data file path
data_reach = pd.read_excel('data\\Final_panel_dataset_reach.xlsx')

filtered_columns = ['crime_reach', 'elevation', 'scaled_betweenness_centrality', 'store_reach', 'lockdown', 'graffiti_reach', 
                    'homeless_shelter_reach', 'traffic_signal_reach', 'street_lighting_poles_reach', 
                    'police_reach', 'transit_reach']

filtered_data_reach = data_reach[filtered_columns]

# Compute the correlation matrix
corr_matrix = filtered_data_reach.corr()

# Select the rows and columns
selected_rows = ['crime_reach', 'transit_reach']  # Adjust these to your actual column names
selected_columns = ['store_reach', 'graffiti_reach', 
                    'homeless_shelter_reach', 'traffic_signal_reach', 'street_lighting_poles_reach', 
                    'police_reach', 'transit_reach']

filtered_matrix_reach = corr_matrix.loc[selected_rows, selected_columns]

# Display the filtered correlation matrix
print(filtered_matrix_reach)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the heatmap for counts
plt.figure(figsize=(10, 6))
sns.heatmap(filtered_matrix_count, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True, cbar_kws={'label': 'Correlation'})
# plt.title('Correlation Matrix for Crime Count and Transit Count')
plt.show()

# Plot the heatmap for reaches
plt.figure(figsize=(10, 6))
sns.heatmap(filtered_matrix_reach, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True, cbar_kws={'label': 'Correlation'})
# plt.title('Correlation Matrix for Crime Reach and Transit Reach')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))

# Plot the heatmap for counts without color bar
sns.heatmap(filtered_matrix_count, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True, cbar=False, ax=axes[0])
axes[0].set_title('Correlation Matrix for Crime Count and Transit Count')

# Plot the heatmap for reaches without color bar
sns.heatmap(filtered_matrix_reach, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True, cbar=False, ax=axes[1])
axes[1].set_title('Correlation Matrix for Crime Reach and Transit Reach')

# Add a single color bar
cbar = fig.colorbar(axes[0].collections[0], ax=axes, orientation='vertical', fraction=0.02, pad=0.04)
cbar.set_label('Correlation')

# Show the plot
plt.show()
```

**Scatter Plot**

```python
# Filter rows to only use 2017
data_count_2017 = data_count[data_count['year'] == 2017]
data_reach_2017 = data_reach[data_reach['year'] == 2017]
```

```python
# Create Variables
crime_count = data_count_2017['crime_count']
transit_count = data_count_2017['transit_count']
store_count = data_count_2017['store_count']
homeless_shelter_count = data_count_2017['homeless_shelter_count']
graffiti_count = data_count_2017['graffiti_count']

crime_reach = data_reach_2017['crime_reach']
transit_reach = data_reach_2017['transit_reach']
store_reach = data_reach_2017['store_reach']
homeless_shelter_reach = data_reach_2017['homeless_shelter_reach']
graffiti_reach = data_reach_2017['graffiti_reach']
```

*Count*

```python
# Create a figure with subplots for scatter plots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Scatter plot 1: Crime count vs Transit count
axes[0, 0].scatter(transit_count, crime_count, alpha=0.7)
axes[0, 0].set_title('Crime Count vs Transit Count')
axes[0, 0].set_xlabel('Transit Count')
axes[0, 0].set_ylabel('Crime Count')

# Scatter plot 2: Crime count vs Store count
axes[0, 1].scatter(store_count, crime_count, alpha=0.7)
axes[0, 1].set_title('Crime Count vs Store Count')
axes[0, 1].set_xlabel('Store Count')
axes[0, 1].set_ylabel('Crime Count')

# Scatter plot 3: Crime count vs Homeless shelter count
axes[1, 0].scatter(homeless_shelter_count, crime_count, alpha=0.7)
axes[1, 0].set_title('Crime Count vs Homeless Shelter Count')
axes[1, 0].set_xlabel('Homeless Shelter Count')
axes[1, 0].set_ylabel('Crime Count')

# Scatter plot 4: Crime count vs Graffiti count
axes[1, 1].scatter(graffiti_count, crime_count, alpha=0.7)
axes[1, 1].set_title('Crime Count vs Graffiti Count')
axes[1, 1].set_xlabel('Graffiti Count')
axes[1, 1].set_ylabel('Crime Count')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

*Reach*

```python
# Create a figure with subplots for scatter plots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Scatter plot 1: Crime reach vs Transit reach
axes[0, 0].scatter(transit_reach, crime_reach, alpha=0.7)
axes[0, 0].set_title('Crime Reach vs Transit Reach')
axes[0, 0].set_xlabel('Transit Reach')
axes[0, 0].set_ylabel('Crime Reach')

# Scatter plot 2: Crime reach vs Store reach
axes[0, 1].scatter(store_reach, crime_reach, alpha=0.7)
axes[0, 1].set_title('Crime Reach vs Store Reach')
axes[0, 1].set_xlabel('Store Reach')
axes[0, 1].set_ylabel('Crime Reach')

# Scatter plot 3: Crime reach vs Homeless shelter reach
axes[1, 0].scatter(homeless_shelter_reach, crime_reach, alpha=0.7)
axes[1, 0].set_title('Crime Reach vs Homeless Shelter Reach')
axes[1, 0].set_xlabel('Homeless Shelter Reach')
axes[1, 0].set_ylabel('Crime Reach')

# Scatter plot 4: Crime reach vs Graffiti reach
axes[1, 1].scatter(graffiti_reach, crime_reach, alpha=0.7)
axes[1, 1].set_title('Crime Reach vs Graffiti Reach')
axes[1, 1].set_xlabel('Graffiti Reach')
axes[1, 1].set_ylabel('Crime Reach')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
